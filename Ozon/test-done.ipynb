__Вопрос:__ Есть ли в данных пропуски? В каких колонках?   
Заполните пропуски пустой строкой для строковых колонок и нулём для числовых.

__Ответ:__ Как видно из вызова метода `info`, пропуски есть только в колонке 'choice_description'. Но на всякий случай напишем решение для всех колонок. 

```python
for column in data:
    data[column] = (data[column].fillna(0) if data[column].dtype == 'int64' else data[column].fillna(''))
data.info()
```

__Вопрос:__ Какая средняя/минимальная/максимальная цена у товара?

```python
data['item_price'] = data['item_price'].str.strip('$').astype('float')
print(f"Средняя цена: ${data['item_price'].mean():.2f}")
print(f"Минимальная цена: ${data['item_price'].min()}")
print(f"Средняя цена: ${data['item_price'].max()}")
```

__Вопрос:__ Разбейте заказы на три группы: дешевые, средние и дорогие. Порог для групп подберите сами (посмотрите, на около какой суммы чаще берут в каждой группе).

__Ответ:__ Визуализируем данные c помощью графика, чтобы понять, как лучше провести разбиение на группы. Для этого предварительно посчитаем сумму в каждом заказе, и выясним, нет ли в данных особенности, мешающих визуализации.

```python
data['order_sum'] = data.quantity * data.item_price
price_count = data[['order_id', 'order_sum']].groupby('order_sum').count()
price_count
```

Наблюдаем выброс - заказ на сумму \\$663.75. Такой заказ серьезно исказит масштаб графика, сделав его нечитаемым. Ограничимся диапазоном цен [0, 150].

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlim([0, 150])
ax.set_xlabel('order price')
ax.set_ylabel('orders amount')
ax.scatter(price_count.index, price_count)
```

Изучив распределение заказов по сумме заказа, разобьем их на 3 группы: до \\$30, от \\$30 до \\$60, от \\$60

```python
cheap = data[data['order_sum'] < 30]
medium = data[(data['order_sum'] >= 30) & (data['order_sum'] < 60)]
expensive = data[data['order_sum'] > 60]
display(cheap, medium, expensive)
```
